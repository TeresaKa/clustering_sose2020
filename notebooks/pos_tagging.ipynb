```python
import pandas as pd
import numpy as np 

import matplotlib.pyplot as plt
import seaborn as sns

import regex as re
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
import nltk
nltk.download('punkt')

import spacy
from spacy import displacy
import en_core_web_sm
```

```python
songs = pd.read_csv("../sample_25percent.csv")
songs["tokenized"] = "0"
songs
```

```python
stop_words = set(stopwords.words('english')) 

def tokenize(text):
    string=re.sub("[^\P{P}]", "", text)
    #word_tokens = word_tokenize(string) 
  
    #filtered_sentence = [w for w in word_tokens if not w in stop_words] 
    return string
```

```python
artist = songs.artist[0]
string = songs.text[0]
string = tokenize(string)
nlp = en_core_web_sm.load()
doc = nlp(string)
```

```python
tokens = []
for token in doc:
    d = {"text":token.text, "lemma":token.lemma_, "pos":token.pos_, "tag":token.tag_, "dep":token.dep_,"shape":token.shape_, "alpha":token.is_alpha, "stop":token.is_stop}
    tokens.append(d)
pos = pd.DataFrame(data=tokens)
```
