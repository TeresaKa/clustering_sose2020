```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
%matplotlib inline
```

### Dataframe laden

```python
df = pd.read_csv("../songtexte_bereinigt_gekuerzt.csv")
df
```

### features (i.d.F. text) vektorisieren 

```python
x = TfidfVectorizer().fit_transform(df.text)
X = TruncatedSVD(n_components=50, random_state=0).fit_transform(x)
```

### PCA anwenden <a name="oneOne"></a>

```python
# 2 Dimensionen
pca = PCA(n_components=2)
```

```python
X_tf = pca.fit_transform(X)
```

```python
x = X_tf[:,0]
y = X_tf[:,1]
plt.figure(figsize=(7,5))
plt.plot(x, y, 'o');
plt.savefig('pca/PCA_Songtexte_gekÃ¼rzt_1.png', bbox_inches = "tight")
```

Label werden als Zahl dargestellt (braucht man manchmal)

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df['labels'] = le.fit_transform(df.artist)
```

```python
df
```

### Dimensionsreduktion wird visualisiert

```python
import numpy as np
import matplotlib.cm as cm
import seaborn as sns

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(frameon=True)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)
plt.subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=0.9,
                wspace=0.0, hspace=0.0)
sns.scatterplot(X_tf[:, 0], X_tf[:, 1],hue=df.labels, marker="o")
plt.title("Songtexte")

plt.savefig('pca/PCA_Songtexte_gekÃ¼rzt_2.png', bbox_inches = "tight")
plt.show()
```

```python
pca.explained_variance_ratio_
```

Die Hauptkomponenten geben die Varianz des Datensatzes an

### Jetzt wird nur ein Teil des Datensatzes verwendet <a name="oneTwo"></a>

```python
from collections import Counter
c = Counter(df.artist).most_common()[20:]
```

```python
sample = df.copy()
```

```python
for el in c:
    sample = sample[sample.artist != el[0]]
sample
```

```python
from sklearn.feature_extraction.text import CountVectorizer

x = TfidfVectorizer(stop_words = None).fit_transform(sample.text)
X = TruncatedSVD(n_components=5, random_state=0).fit_transform(x)
```

```python
# 3 Dimensionen
pca = PCA(n_components=3)
X_tf = pca.fit_transform(X)
```

```python
x = X_tf[:,0]
y = X_tf[:,1]
plt.figure(figsize=(7,5))
plt.plot(x, y, 'o');
```

```python
import numpy as np
import matplotlib.cm as cm
import seaborn as sns

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(frameon=True)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)
plt.subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=0.9,
                wspace=0.0, hspace=0.0)
sns.scatterplot(X_tf[:, 0], X_tf[:, 1],hue= sample['labels'], marker="o", legend=False)

plt.savefig('pca/PCA_Songtexte_gekÃ¼rzt_3_20_hÃ¤ufigsten.png', bbox_inches = "tight")
plt.show()
```

```python
pca.explained_variance_ratio_
```

### MDS - Multidimensionale Skalierung

```python
from sklearn import manifold
model = manifold.MDS(2, max_iter=100, n_init=1)
X_tf = model.fit_transform(X)

import numpy as np
import matplotlib.cm as cm
import seaborn as sns

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(frameon=True)
ax.set_xlabel('Component 1', fontsize = 15)
ax.set_ylabel('Component 2', fontsize = 15)
ax.set_title('MSD', fontsize = 20)
plt.subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=0.9,
                wspace=0.0, hspace=0.0)
sns.scatterplot(X_tf[:, 0], X_tf[:, 1],hue= sample['labels'], marker="o")

plt.savefig('pca/MDS_Songtexte_gekÃ¼rzt_20_hÃ¤uifgsten.png', bbox_inches = "tight")
plt.show()

```

### verkleinerter Datensatz wird angewendet <a name="oneThree"></a>

```python
df = pd.read_csv("../songs_25.csv")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

x = TfidfVectorizer(stop_words = None).fit_transform(df.text)
X = TruncatedSVD(n_components=5, random_state=0).fit_transform(x)
```

### PCA mit POS

```python
df = pd.read_csv("../songs_25.csv")
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

x = TfidfVectorizer().fit_transform(df.POS.values.astype(str))
X = TruncatedSVD(n_components=5, random_state=0).fit_transform(x)
```
