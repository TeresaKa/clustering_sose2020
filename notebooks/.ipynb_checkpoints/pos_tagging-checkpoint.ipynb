```python
import pandas as pd
import numpy as np 

import matplotlib.pyplot as plt
import seaborn as sns

import regex as re
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
import nltk
nltk.download('punkt')

import spacy
from spacy import displacy
import en_core_web_sm
```

```python
songs = pd.read_csv("songs_plus_dates.csv")
songs["POS"] = "0"
songs
```

### use only nouns and adjectives

```python
nlp = en_core_web_sm.load()
tokens = []
for i, row in songs.iterrows():
    doc = nlp(row.text)
    for token in doc:
        d = {"text":token.text, "lemma":token.lemma_, "pos":token.pos_, "tag":token.tag_, "dep":token.dep_,"shape":token.shape_, "alpha":token.is_alpha, "stop":token.is_stop}
        tokens.append(d)
    pos = pd.DataFrame(data=tokens)
    tokens = []  
    x = list(pos[pos.tag=="JJ"].text) + list(pos[pos.tag=="NN"].text) + list(pos[pos.tag=="VBP"].text)
    x = ' '.join(x)
    songs.at[i,'POS'] = x
#     songs.at[i, 'nouns'] = len(pos[pos.tag=="NN"].text)
#     songs.at[i, 'adjectives'] = len(pos[pos.tag=="JJ"].text)
#     songs.at[i, 'verb'] = len(pos[pos.tag=="VBP"].text)    

songs
```
